```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/covid-data.csv")
df
```

```python
df1=df.query("location=='India'")
df1
```

```python
fig=px.bar(df1,x='date',y='total_cases')
fig.show()
```

```python
fig=px.line(df1,x='date',y='total_cases')
fig.show()
```

```python
fig=px.bar(df1,x='date',y='total_deaths',color="total_cases")
fig.show()
```

```python
fig=px.line(df1,x='total_cases',y='location',color="total_deaths",animation_frame='date')
fig.show()
```

```python
fig=px.pie(df,values='total_cases',names='location',color='total_deaths')
fig.show()
```

```python
fig=px.pie(df1,values='total_cases',names='location',color='total_deaths',hole=0.5)
fig.show()
```

```python
fig=px.scatter_3d(df1,x='total_cases',y='total_deaths',z='date',color='location')
fig.show()
```
